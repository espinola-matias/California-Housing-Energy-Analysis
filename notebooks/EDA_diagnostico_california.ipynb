Conteo de valores de columnas faltantes, identificacion y sacar un porcentaje de ello

```python
# Sumamos los valores nulos (True = 1) en cada columna
missing_counts = df_original.isnull().sum()

# Calculamos el porcentaje de nulos
total_rows = len(df_original)
missing_percentage = (missing_counts / total_rows) * 100

# Creamos un DataFrame conciso para visualizar rapidamente
missing_info = pd.DataFrame({
    'Total Faltantes': missing_counts,
    'Porcentaje (%)': missing_percentage.round(2)
})

# Filtramos solo las columnas que tienen algun valor faltante
missing_info = missing_info[missing_info['Total Faltantes'] > 0]

print(missing_info)
```
